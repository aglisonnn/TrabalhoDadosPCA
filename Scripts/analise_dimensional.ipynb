```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# 1. Carregar os dados
df_scaled = pd.read_csv('../dados_processados/features_clientes_padronizadas.csv')

# 2. Filtrar: remover IDs e manter apenas números
# IMPORTANTE: Adicionei o drop do CustomerID
df_numeric = df_scaled.drop(columns=['CustomerID'], errors='ignore').select_dtypes(include=['number'])

# 3. Aplicar o PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(df_numeric)
df_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])

# 4. Mostrar a Variância (Agora deve ser diferente de 100/0)
var_exp = pca.explained_variance_ratio_
print(f"Variância Explicada PC1: {var_exp[0]:.2%}")
print(f"Variância Explicada PC2: {var_exp[1]:.2%}")

# 5. GERAR O GRÁFICO (O que faltava aparecer!)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', alpha=0.7)
plt.title('Projeção PCA 2D dos Clientes')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
